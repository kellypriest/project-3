```python
top_20_breeds = pd.DataFrame(breed_counts2.head(20))
top_breeds = top_20_breeds.reset_index(drop = False)
top_breeds = top_breeds.rename(columns={"index": "PrimaryBreed", "PrimaryBreed": "count"})
top_breeds
```

```python
#make breed outcome counts csv
a_count = []
t_count = []
f_count = []
e_count = []

for index, row in top_breeds.iterrows():
    breed = row[0]  
    a_sum = 0
    t_sum = 0
    f_sum = 0
    e_sum = 0
    
    for index2, row2 in df.iterrows(): 
        
        if row2[3] == breed:
            outcome = row2[1]
          
            if outcome == 'ADOPTION':
                a_sum += 1
            elif outcome == 'EUTHANASIA':
                e_sum +=1
            elif outcome == 'TRANSFER':
                t_sum +=1
            else:
                f_sum +=1
    a_count.append(a_sum)
    e_count.append(e_sum)
    t_count.append(t_sum)
    f_count.append(f_sum)
            
           
a_count      
    
```

```python
breeds_df = top_breeds
breeds_df['Adoption'] = a_count
breeds_df['Euthanasia'] = e_count
breeds_df['Transfer'] = t_count
breeds_df['Foster'] = f_count
breeds_df
```

```python
#make gender outcome types csv
a_count = []
t_count = []
f_count = []
e_count = []
gender = ['Female', 'Male']

for g in gender:
    a_sum = 0
    t_sum = 0
    f_sum = 0
    e_sum = 0
    
    for index, row in df.iterrows(): 
    
        if row[5] == g:
            outcome = row[1]
            if outcome == 'ADOPTION':
                a_sum += 1
            elif outcome == 'EUTHANASIA':
                e_sum +=1
            elif outcome == 'TRANSFER':
                t_sum +=1
            else:
                f_sum +=1

    a_count.append(a_sum)
    e_count.append(e_sum)
    t_count.append(t_sum)
    f_count.append(f_sum)
            
```

```python
gender_df = pd.DataFrame(list(zip(gender, a_count)), 
               columns =['Sex', 'Adoption']) 
gender_df['Euthanasia'] = e_count
gender_df['Transfer'] = t_count
gender_df['Foster'] = f_count
gender_df
```

```python
#make age outcome count csv
a_count = []
t_count = []
f_count = []
e_count = []
age_cate = ['Baby', 'Young', 'Adult', 'Senior']

for age in age_cate:
    a_sum = 0
    t_sum = 0
    f_sum = 0
    e_sum = 0
    
    for index, row in df.iterrows(): 
    
        if row[2] == age:
            outcome = row[1]
            if outcome == 'ADOPTION':
                a_sum += 1
            elif outcome == 'EUTHANASIA':
                e_sum +=1
            elif outcome == 'TRANSFER':
                t_sum +=1
            else:
                f_sum +=1

    a_count.append(a_sum)
    e_count.append(e_sum)
    t_count.append(t_sum)
    f_count.append(f_sum)
            
```

```python
age_group_df = pd.DataFrame(list(zip(age_cate, a_count)), 
               columns =['Age Group', 'Adoption']) 
age_group_df['Euthanasia'] = e_count
age_group_df['Transfer'] = t_count
age_group_df['Foster'] = f_count
age_group_df
```
